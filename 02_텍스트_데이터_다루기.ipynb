데이터 가져오기

```python
import csv # 기본적으로 일반 파이썬에선 CSV파일이 열리지 않는다.

with open("/content/Demographic_Statistics_By_Zip_Code.csv") as infile:
  data = list(csv.reader(infile))
```

```python
# 첫번째 레코드에 COUNT PARTICIPANTS 데이터 인덱스 값 추출 data[0] 컬럼값
countParticipantsindex = data[0].index("COUNT PARTICIPANTS")
countParticipantsindex
# 데이터가 나중에 데이터를 정제하거나 수정했을 때 특정 컬럼위 위치가 바뀔 수 있기 때문에
# countParticipantsindex에 컬럼 인덱스를 넣어두는 것
```

```python
# COUNT PARTICIPANTS의 데이터 추출

countParticipants = [int(row[countParticipantsindex]) for row in data[1:]] # 리스트 내포
# 0번째는 컬럼 정보가 들어있기 때문에 가져오지 않는다.
# 통계적으로 계산하려면 int를 이용해서 숫자타입으로 바꿔줘야 한다.
# 암묵적인 형 변환을 믿지 마라, 명시적으로 해줘야 한다. 컴퓨터를 믿으면 안된다. 
countParticipants
```

나이의 평균과 표준편차 구하기

```python
import statistics

print(statistics.mean(countParticipants), statistics.stdev(countParticipants)) 
```

```python
nltk.download('wordnet')
```

```python
wn = nltk.corpus.wordnet  # 코퍼스 리더 (reader)선언

# synset : 유의어 묶음으로 단어-품사-순번 으로 구성
wn.synsets("cat")  # cat과 유의어 묶음 목차를 봤다 생각
```

```python
wn.synset('cat.n.01').hypernyms()  # hypernyms는 상위어
wn.synset('cat.n.01').hyponyms()  # hyponyms는 하위어

# domestic_cat 집고양이 wildcat 길고양이
```

wordNet 사용해서 synset 간 의미론적 유사도 계산하기 (0-1사이 실수)

1.   0 이면 두 단어 서로 관계 없음
2.   1 이면 완전한 유의어

단어가 나왔을 때 이 것을 사전적인 의미로만 긍정 부정으로 나눌 지,가 연구한 긍정 부정으로 나눌 지 확인하기위해 단어들이 어떤 관련이 있는지 확인해봐야 함



```python
# 고양이, 링스 ; 시라소니 유사도 계산
x = wn.synset('cat.n.01') 
y = wn.synset('lynx.n.01')

# x,y 유사도 계산
x.path_similarity(y)  #0.04 유사성이 거의 없다.
```

```python
# 고양이 ; 집고양이 유사도 계산
x = wn.synset('cat.n.01') 
y = wn.synset('domestic_cat.n.01')

# x,y 유사도 계산
x.path_similarity(y)
```

```python
# 고양이 ; 고양이 유사도 계산
x = wn.synset('cat.n.01') 
y = wn.synset('cat.n.01')

# x,y 유사도 계산
x.path_similarity(y)

# 유사도 계산은 wordnet에서 연구한 결과로 나오는 것이다.
```

```python
# 강아지 : 고양이 유사도 계산
x = wn.synset('cat.n.01') 
y = wn.synset('dog.n.01')
# x,y 유사도 계산
x.path_similarity(y)
```

```python
# www.networksciencelab.com 사이트에서 하이퍼 링크 정보 모두 가져오기
from bs4 import BeautifulSoup
from urllib.request import urlopen

with urlopen("http://www.networksciencelab.com/") as doc:
  soup = BeautifulSoup(doc)
# http://가 없으면 통신을 할 수가 없다.


# 하이퍼링크만 추출하기  # 튜플리스트로 가져옴
links = [(link.string, link["href"]) for link in soup.find_all("a") if link.has_attr("href")] 
# a 태그 안에 항상 링크 정보가 들어있다.

# 책 제목만 추출
html_text = [i[0] for i in links]
# html_text에는 리스트 형태로 저장되게 된다.

# 0,1번인덱스에 있는 none데이터 제거하여 -> 리스트에서 하나의 텍스트로 변환
html_text = " ".join(html_text[2:])
# 우리는 텍스트 분석을 하고 싶은 것이므로 하나의 문장으로 된 텍스트가 필요하다.(리스트x)
# 때문에 join함수를 사용하여 리스트 안에 있는 모든 요솟값들을 하나의 문장으로 연결했다.
html_text
```

index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
from bs4 import BeautifulSoup
from collections import Counter # 많이 쓰이는 단어를 세기 위해
from nltk.corpus import stopwords  # 불용어들을 삭제하기 위해
from nltk import LancasterStemmer  # 원형 데이터를 찾아내기 위해

# 형태소 분류기를 생성한다.
ls = nltk.LancasterStemmer()

# 파일을 읽고 soup을 만든다.
with open("/content/index.html") as infile:
  soup = BeautifulSoup(infile)  # 통상 적인 텍스트 파일이 아니라 html파일이기 때문에 BeautifulSoup을 통해 불러들여야 한다.
  
```

```python
# 텍스트 추출 -> 토큰으로 만들어야 한다.(토큰화)
# 절대!! 변수를 a b c 이런식으로 지으면 안된다. 인지하고 알기 쉽게 작명해야 한다.
# 코딩 책에 나오는 스타일처럼 변수명을 작명해야 한다.
# 변수 선언할 때 무조건 소문자. 단어끼리 연결되어 있다면 뒤에 단어 첫단어는 대문자
# 변하지 않는 불변의 값은 대문자로 쓴다.
# 여러개의 요소들이 들어있는 경우 경우 s를 붙여 복수형으로 변수를 만든다.
# 1) 정형화를 시작할 때 첫번째 단계는 토큰화
words = nltk.word_tokenize(soup.text)

words
```

```python
# 2) 단어를 소문자로 변환
words = [w.lower() for w in words]  # 리스트 내포
words
```

```python
import nltk
nltk.download('stopwords')
```

```python
# 3) 불용어를 제거하고 단어의 형태소를 추출
# (불용어 제거는 stopwords.words("english")
# 특수기호 제거는 isalnum() : isalnum() 문자열이 알파벳([a-zA-Z])과 숫자([0-9])로만 구성
# isalnum 숫자 형태를 가져오는 것 -> not in이니까 숫자는 버리는것
words = [ls.stem(w) for w in words if w not in stopwords.words("english") and w.isalnum()]  

```

```python
print(words)
```

```python
# 4) 가장 빈번하게 등장하는 단어 10개 추출
# freq 변수는 빈도 얘기할 때 많이 쓴다.
freqs = Counter(words)
print(freqs.most_common(5))

# [('network', 16), ('soc', 8), ('onlin', 7), ('inform', 4), ('gam', 4)] 
# -> network관련한 단어가 가장 많이 나온다는 것을 알 수 있다.
```
